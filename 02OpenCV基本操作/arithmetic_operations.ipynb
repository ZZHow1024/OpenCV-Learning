# 2.2算数操作
- 导入相关库

```python
import cv2 as cv
import matplotlib.pyplot as plt
```

- 读入两张图像

```python
image = cv.imread('images/image.jpg', cv.IMREAD_COLOR)
raindrops = cv.imread('images/raindrops.jpg', cv.IMREAD_COLOR)
```

## 图像的加法
- OpenCV 加法

```python
image_raindrops = cv.add(image, raindrops)
plt.imshow(image_raindrops[:, :, ::-1])
```

- NumPy 加法

```python
image_raindrops = image + raindrops
plt.imshow(image_raindrops[:, :, ::-1])
```

## 图像的混合
- 混合公式：
$dist = \alpha ⋅ image_1 + \beta ⋅ image_2 + \gamma$
- 第一幅图的权重为 0.3，第二幅图的权重为 0.7，$\gamma$ 值为 0

```python
image_raindrops = cv.addWeighted(image, 0.3, raindrops, 0.7, 0)
plt.imshow(image_raindrops[:, :, ::-1])
```
