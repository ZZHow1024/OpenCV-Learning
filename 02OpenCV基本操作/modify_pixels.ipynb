# 获取并修改图像中的像素点
- 导入相关库

```python
import numpy as np
import matplotlib.pyplot as plt
```

- 创建一个 256 * 256 的纯黑图像

```python
image = np.zeros((256, 256, 3), np.uint8)
plt.imshow(image[:, :, ::-1])
```

- 获取某个像素点的像素值

```python
image[100, 100]
```

- 仅获取蓝色通道的强度值

```python
image[100, 100, 0]
```

- 修改某个位置的像素值

```python
image[100, 100] = (0, 0, 255)
plt.imshow(image[:, :, ::-1])
```
