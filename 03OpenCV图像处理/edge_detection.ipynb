# 边缘检测

- 导入相关库

```python
import cv2 as cv
import zhplot
import matplotlib.pyplot as plt

zhplot.matplotlib_chineseize()
```

- 展示原图

```python
image = cv.imread('./images/image.jpg', cv.IMREAD_GRAYSCALE)
plt.title('原图')
plt.imshow(image, cmap='gray')
```

## Sobel 算子

```python
x = cv.Sobel(image, cv.CV_16S, 1, 0)
y = cv.Sobel(image, cv.CV_16S, 0, 1)
abs_x = cv.convertScaleAbs(x)
abs_y = cv.convertScaleAbs(y)
res = cv.addWeighted(abs_x, 0.5, abs_y, 0.5, 0)
plt.title('Sobel 算子')
plt.imshow(res, cmap='gray')
```

## Scharr 算子

```python
x = cv.Sobel(image, cv.CV_16S, 1, 0, ksize=-1)
y = cv.Sobel(image, cv.CV_16S, 0, 1, ksize=-1)
abs_x = cv.convertScaleAbs(x)
abs_y = cv.convertScaleAbs(y)
res = cv.addWeighted(abs_x, 0.5, abs_y, 0.5, 0)
plt.title('Scharr 算子')
plt.imshow(res, cmap='gray')
```

## Laplacian 算子

```python
res = cv.Laplacian(image, cv.CV_16S)
res = cv.convertScaleAbs(res)
plt.title('Laplacian 算子')
plt.imshow(res, cmap='gray')
```

## Canny 边缘检测

```python
res = cv.Canny(image, 100, 200)
plt.title('Canny 边缘检测')
plt.imshow(res, cmap='gray')
```
