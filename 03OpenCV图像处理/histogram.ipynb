# 直方图

- 导入相关库

```python
import numpy as np
import cv2 as cv
import zhplot
import matplotlib.pyplot as plt

zhplot.matplotlib_chineseize()
```

- 展示原图

```python
image = cv.imread('./images/image.jpg')
plt.title('原图')
plt.imshow(image[:, :, ::-1])
```

## 灰度直方图

```python
hist = cv.calcHist([image], [0], None, [256], [0, 256])
plt.figure(figsize=(10, 8))
plt.title('B 通道灰度直方图')
plt.plot(hist)
```

## 掩膜的应用

```python
# 创建掩膜
mask = np.zeros(image.shape[:2], np.uint8)
mask[70:210, 195:395] = 1
plt.title('掩膜')
plt.imshow(mask, cmap='gray')
```

```python
mask_image = cv.bitwise_and(image, image, mask=mask)
plt.title('图像掩膜')
plt.imshow(mask_image, cmap='gray')
```

```python
hist = cv.calcHist([image], [0], mask, [256], [0, 256])
plt.figure(figsize=(10, 8))
plt.title('图像掩膜后 B 通道灰度直方图')
plt.plot(hist)
```

## 直方图均衡化

```python
image = cv.imread('./images/image.jpg', cv.IMREAD_GRAYSCALE)
plt.title('灰度原图')
plt.imshow(image, cmap='gray')
```

```python
res = cv.equalizeHist(image)
plt.title('直方图均衡化')
plt.imshow(res, cmap='gray')
```

## 自适应的直方图均衡化

```python
cl = cv.createCLAHE(40, (8, 8))
res = cl.apply(image)
plt.title('自适应的直方图均衡化')
plt.imshow(res, cmap='gray')
```
