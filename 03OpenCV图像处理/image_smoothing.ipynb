# 图像平滑

- 导入相关库

```python
import cv2 as cv
import zhplot
import matplotlib.pyplot as plt

zhplot.matplotlib_chineseize()
```

- 展示原图

```python
image = cv.imread('./images/image.jpg')
plt.title('原图')
plt.imshow(image[:, :, ::-1])
```

## 均值滤波

```python
res = cv.blur(image, (10, 10))
plt.title('均值滤波')
plt.imshow(res[:, :, ::-1])
```

## 高斯滤波

```python
res = cv.GaussianBlur(image, (5, 5), 1)
plt.title('高斯滤波')
plt.imshow(res[:, :, ::-1])
```

## 中值滤波

```python
res = cv.medianBlur(image, 3)
plt.title('中值滤波')
plt.imshow(res[:, :, ::-1])
```
