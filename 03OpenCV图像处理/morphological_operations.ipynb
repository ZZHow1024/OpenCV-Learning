# 形态学操作

- 导入相关库

```python
import numpy as np
import cv2 as cv
import zhplot
import matplotlib.pyplot as plt

zhplot.matplotlib_chineseize()
```

- 展示原图

```python
image = cv.imread('./images/image.jpg')
plt.title('原图')
plt.imshow(image[:, :, ::-1])
```

## 腐蚀与膨胀

```python
kernel = np.ones((5, 5), np.uint8) # 创建核结构
```

```python
res = cv.erode(image, kernel, iterations=1) # 腐蚀 1 次
plt.title('腐蚀 1 次')
plt.imshow(res[:, :, ::-1])
```

```python
res = cv.dilate(image, kernel, iterations=1) # 膨胀 1 次
plt.title('膨胀 1 次')
plt.imshow(res[:, :, ::-1])
```

## 开闭运算

```python
res = cv.morphologyEx(image, cv.MORPH_OPEN, kernel) # 开运算
plt.title('开运算')
plt.imshow(res[:, :, ::-1])
```

```python
res = cv.morphologyEx(image, cv.MORPH_CLOSE, kernel) # 闭运算
plt.title('闭运算')
plt.imshow(res[:, :, ::-1])
```

## 礼帽和黑帽

```python
res = cv.morphologyEx(image, cv.MORPH_TOPHAT, kernel) # 礼帽运算
plt.title('礼帽运算')
plt.imshow(res[:, :, ::-1])
```

```python
res = cv.morphologyEx(image, cv.MORPH_BLACKHAT, kernel) # 黑帽运算
plt.title('黑帽运算')
plt.imshow(res[:, :, ::-1])
```
