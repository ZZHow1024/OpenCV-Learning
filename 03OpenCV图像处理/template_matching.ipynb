# 模版匹配

- 导入相关库

```python
import cv2 as cv
import zhplot
import matplotlib.pyplot as plt

zhplot.matplotlib_chineseize()
```

- 原图展示

```python
image = cv.imread('./images/image.jpg', cv.IMREAD_COLOR)
plt.title('原图')
plt.imshow(image[:, :, ::-1])
```

- 模版展示

```python
template = cv.imread('./images/dog_head.png', cv.IMREAD_COLOR)
plt.title('模版展示')
plt.imshow(template[:, :, ::-1])
```

- 开始模版匹配

```python
res = cv.matchTemplate(image, template, cv.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv.minMaxLoc(res)
top_left = max_loc
height, width = template.shape[:2]
bottom_right = (top_left[0] + width, top_left[1] + height)
cv.rectangle(image, top_left, bottom_right, (0, 0, 255), 2)
plt.title('模版匹配结果')
plt.imshow(image[:, :, ::-1])
```
